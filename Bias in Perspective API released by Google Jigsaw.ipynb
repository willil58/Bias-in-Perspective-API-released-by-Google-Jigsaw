```python
from collections import Counter
```

```python
def find_most_common_words_above_threshold(df, thresh):
    df_subset = df[df['score']>thresh]
    text = ' '.join(df_subset['comment_text'].values)
    split_text = text.split()
    counter = Counter(split_text)
    print(counter)
```
